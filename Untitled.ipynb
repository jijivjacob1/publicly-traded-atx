```python
import pandas as pd
import numpy as np

from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
from sqlalchemy import inspect

from flask import (
    Flask,
    render_template,
    jsonify)
```

```python
engine = create_engine("sqlite:///pub_atx.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
cmpny = Base.classes.cmpny
```

```python
sel = [cmpny.id_cmpny ,
    cmpny.rank ,
    cmpny.name,
    cmpny.address ,
    cmpny.city ,
    cmpny.state ,
    cmpny.zip_code,
    cmpny.phn_nbr,
    cmpny.website ,
    cmpny.tckr ,
    cmpny.ipo_yr,
    cmpny.yr_founded ,
    cmpny.gen_buss_desc ,
    cmpny.curr_top_exec,
    cmpny.yr_estblsh ,
    cmpny.austin_staff_cnt ,
    cmpny.comp_staff_cnt ,
    cmpny.tot_local_emp_cnt ,
    cmpny.exchng ,
    cmpny.lat ,
    cmpny.lng ,
    cmpny.industry ,
    cmpny.sector ,
    cmpny.marketcap ]
```

```python
inspector = inspect(engine)
```

```python
inspector
```

```python
session = Session(engine)
```

```python
app = Flask(__name__)
```

```python



results = session.query(*sel).all()

df = pd.DataFrame(results, columns=[*sel])
df
```

```python
# create route that renders index.html template
@app.route("/")
def home():
    return render_template("index.html")
```
